## ğŸ”¹ Part 1: Insertion by Name (Sorted Order)
I have  students list of dictionaries,
  First I do insertion sort of the list  by the `"name"` key (alphabetically). This is done using a simple linear scan to find the correct position to insert.

It builds the sorted list one item at a time. It picks an element and inserts it into its correct position in the already sorted part of the list.
# Procedure 
1, Start with the second item in the list (since the first one is already "sorted" by itself).
2, Compare the current item with the items before it.
3, If the current item is smaller (alphabetically or numerically), shift the previous item to the right.
4, Keep moving left until you find the right place for the current item.
5, Insert the item at that position.
6, Repeat for all remaining items in the list


```python

# Insertion Sort by name
# Inserted student data
students = [
    {"name": "Yordi", "cgpa": 3.5, "department": "CS", "id": 1601896},
    {"name": "Beti", "cgpa": 3.9, "department": "IT", "id": 1601232},
    {"name": "Zenawt", "cgpa": 2.8, "department": "DS", "id": 1601111},
    {"name": "Dawit", "cgpa": 4.0, "department": "IS", "id": 1601567},
    {"name": "Eden", "cgpa": 3.1, "department": "SE", "id": 1601678},
    {"name": "Addis", "cgpa": 3.09, "department": "SE", "id": 1601898},
    {"name": "Tewbu", "cgpa": 3.21, "department": "IT", "id": 16017678},
    {"name": "Hlina", "cgpa": 3.1, "department": "IT", "id": 1601690},
    {"name": "Hana", "cgpa": 3.53, "department": "CS", "id": 1601634},
    {"name": "TSEGA", "cgpa": 3.91, "department": "ME", "id": 1601458},
    {"name": "Solomon", "cgpa": 3.35, "department": "CS", "id": 1601612},
]


for i in range(1, len(students)):
    key = students[i]
    j = i - 1
    while j >= 0 and key["name"] < students[j]["name"]:
        students[j + 1] = students[j]
        j -= 1
    students[j + 1] = key

print("After insertion sort by name:")
for s in students:
    print(s)
```


### ğŸ”¹ Part 1: Selection by CGPA  (Sorted Order)
I have students data and i do the selection sort of the data by **CGPA**

It repeatedly finds the smallest (or largest) element and places it at the correct position by swapping it
  # Procedure 
1, Start at the first position.
2, Look through the entire list to find the smallest item (e.g., lowest CGPA).
3, Swap it with the item at the first position.
4, Move to the second position and repeat the same process for the remaining list.
5, Keep repeating this until all items are in order.

```python
# Selection Sort by CGPA

students = [
    {"name": "Yordi", "cgpa": 3.5, "department": "CS", "id": 1601896},
    {"name": "Beti", "cgpa": 3.9, "department": "IT", "id": 1601232},
    {"name": "Zenawt", "cgpa": 2.8, "department": "DS", "id": 1601111},
    {"name": "Dawit", "cgpa": 3.7, "department": "IS", "id": 1601567},
    {"name": "Eden", "cgpa": 3.1, "department": "SE", "id": 1601678},
    {"name": "Addis", "cgpa": 3.09, "department": "SE", "id": 1601898},
    {"name": "Tewbu", "cgpa": 3.21, "department": "IT", "id": 16017678},
    {"name": "Hlina", "cgpa": 3.1, "department": "IT", "id": 1601690},
    {"name": "Hana", "cgpa": 3.53, "department": "CS", "id": 1601634},
    {"name": "TSEGA", "cgpa": 3.91, "department": "ME", "id": 1601458},
    {"name": "Somon", "cgpa": 3.35, "department": "CS", "id": 1601612},
]
for i in range(len(students)):
    min_index = i
    for j in range(i + 1, len(students)):
        if students[j]["cgpa"] < students[min_index]["cgpa"]:
            min_index = j
    students[i], students[min_index] = students[min_index], students[i]

print("\nAfter selection sort by CGPA:")
for s in students:
    print(s)
```

# Bubble sort by students id
It compares each pair of adjacent items and swaps them if theyâ€™re in the wrong order. This repeats until the list is sorted.
ğŸ”· Procedure :
1, Start from the first item and compare it with the next item.
2, If the first item is greater than the second, swap them.
3, Move to the next pair and repeat the comparison and swap if needed.
4, Continue to the end of the list; the largest item will "bubble up" to the end.
5, Repeat the process for the rest of the list (ignoring the last sorted items).
6, Stop when no more swaps are needed in a full pass.


```python

# Bubble Sort by ID


students = [
    {"name": "Yordi", "cgpa": 3.5, "department": "CS", "id": 1601896},
    {"name": "Beti", "cgpa": 3.9, "department": "IT", "id": 1601232},
    {"name": "Zenawt", "cgpa": 2.8, "department": "DS", "id": 1601111},
    {"name": "Dawit", "cgpa": 3.7, "department": "IS", "id": 1601567},
    {"name": "Addis", "cgpa": 3.09, "department": "SE", "id": 1601898},
    {"name": "Tewbu", "cgpa": 3.21, "department": "IT", "id": 16017678},
    {"name": "Hlina", "cgpa": 3.1, "department": "IT", "id": 1601690},
    {"name": "Hana", "cgpa": 3.53, "department": "CS", "id": 1601634},
    {"name": "TSEGA", "cgpa": 3.91, "department": "ME", "id": 1601458},
    {"name": "Somon", "cgpa": 3.35, "department": "CS", "id": 1601612},
]

for i in range(len(students)):
    for j in range(0, len(students) - i - 1):
        if students[j]["id"] > students[j + 1]["id"]:
            students[j], students[j + 1] = students[j + 1], students[j]

print("\nAfter bubble sort by ID:")
for s in students:
    print(s)
```
